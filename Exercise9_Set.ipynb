```python
#Creating a set
color={"Red", "Green", "Yellow", "Black"}
print(color)
```

```python
#Creating a set using set function
n=set((2,1,3,2,1))
print(n)
```

```python
#Basic set operation
a={1,2,3}
b={3,4,5}
print(a.union(b))
print(b.union(a))
print(a.intersection(b))
print(a.difference(b))
print(b.difference(a))
print(a.symmetric_difference(b))
print(b.symmetric_difference(a))
```

```python
#Adding a new element in a set
os={"Windows", "Linux","MacOS"}
os.add("Unix")
print(os)
```

```python
#Subject chosen
maths={"Sureka","Ranjana","Kaniga","Tharshini","Charunethra"}
cs={"Dhanya","Ranjana","Alfhiya","Alshifa","Kaniga"}
print("Students who are chosen both subjects:",maths.intersection(cs))
print("Students who are chosen either maths or cs:",maths.symmetric_difference(cs))
print("Students who are chosen maths alone:",maths.difference(cs))
print("Students who are chosen CS alone:",cs.difference(maths))
maths.add("Riya")
print(maths)
```

```python
#Updating a set
maths={"Sureka","Ranjana","Kaniga","Tharshini","Charunethra"}
m={"Meena", "Reena"}
maths.update(m)
print(maths)
```

```python
#Type error Due to unhashable set element
maths={"Sureka", "Ranjana", "Kaniga", "Alfhiya", "Alshifa"}
maths.add({"Theju","Anitha","Reka"})
print(maths)
```

```python
#Discard function
maths={"Sureka", "Ranjana", "Kaniga", "Alfhiya", "Alshifa"}
maths.discard("Sureka")
print(maths)
```

```python
#Remove function
maths={"Sureka", "Ranjana", "Kaniga", "Alfhiya", "Alshifa"}
maths.remove("Ranjana")
print(maths)
```

```python
#Key error while removing non-existent element from set
maths={"Sureka", "Ranjana", "Kaniga", "Alfhiya", "Alshifa"}
maths.discard("Sureka")
maths.remove("Sureka")
print(maths)
```

```python
#Avoiding duplicates
s={'a','c','e','d'}
s.update('and')
print(s)
```

```python
#Clear function
s={'a','c','e','d'}
s.clear()
print(s)
```

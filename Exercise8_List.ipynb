```python
#Creating a list 
numbers=[2,4,6,8,10]
colors=["Red","blue","pink","yellow","black"]
combined_list=[109,80.9,"Sureka",True]
empty_list=[]
print(numbers,colors,combined_list,empty_list)
```

```python
#Accessing list elements
colors=["Red","blue","pink","yellow","black"]
print(colors[0])
print(colors[-5])
print(colors[4])
print(colors[-1])
```

```python
#Modify a list
colors=["Red","blue","pink","yellow","black"]
colors[1]="Lavender"
colors[2]="Green"
print(colors)
```

```python
#Common list operations-01
colors=["Red","blue","pink","yellow","black"]
colors.append("White")
colors.insert(1,"Orange")
print(colors)
```

```python
#Common list operations-02
colors=["Red","blue","pink","yellow","black"]
colors.pop()
print(colors)
```

```python
#Common list operations-03
colors=["Red","blue","pink","yellow","black"]
colors.remove("Red")
print(colors)
```

```python
#Common list operations-04
colors=["Red","blue","pink","yellow","black"]
combined_list=[109,80.9,"Ashifa",True]
print(len(colors))
print(len(combined_list))
```

```python
#Build-in functions in list
a=[5,34,98,20,25]
print(min(a))
print(max(a))
a.sort()
print(a)
a.reverse()
print(a)
```

```python
#Looping in list
colors=["Red","blue","pink","yellow","black"]
for i in colors:
    print(i)
```

```python
#List slicing
a=[5,34,98,20,25]
print(a[0:2])
print(a[::-1])
```

```python
#Square of a number from ranges 1 to 5
s=[n**2 for n in range(1,6)]
print(s)
```

```python
#Square of numbers
l1=[5,2,4,7,8]
s=[n**2 for n in l1]
print(s)
```

```python
#Count how many numbers are even in the list
z=[3,8,9,10,45,22]
w=0
for n in z:
    if n%2==0:
        w+=1
print(w)
```
